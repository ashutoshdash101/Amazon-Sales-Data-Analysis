### Changing the order and ship date type to date_time format

```python
data['Order Date']=pd.to_datetime(data['Order Date'])
data['Ship Date']=pd.to_datetime(data['Ship Date'])
```

```python
data.info()
```

```python
data.head()
```

### Adding Year and Month column from the Order Date column to the dataset 


```python
data['Year'] = data['Order Date'].dt.year
data['Month'] = data['Order Date'].dt.month
data['Month_Year'] = data['Order Date'].dt.strftime('%Y-%m')
```

```python
data.head()
```

## <b>Month Wise Sales Trend<b>

```python
monthly_sales = data.groupby('Month')['Total Revenue'].sum()
```

```python
# lineplot representation of the month wise sales
sns.barplot(x=monthly_sales.index, y=monthly_sales.values)
plt.title("Monthly Wise Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.xticks(np.arange(0, 13, step=1))
plt.show()
```

* The company reaches highest revenue in the February month and the lowest in the August month

## <b>Year wise Sales trend<b>

```python
yearly_sales = data.groupby('Year')['Total Revenue'].sum()
```

```python
sns.lineplot(x=yearly_sales.index, y=yearly_sales.values, color='red',
             marker='o')
plt.title("Yearly Wise Sales Trend")
plt.xlabel("Year")
plt.ylabel("Sales")
# plt.xticks(np.arange(0, 13, step=1))
plt.show()
```

* The company made a huge sale in the year 2012 and a much less in the year 2011.
* After the year 2012 the rate of sales decreases and increased a little from 2015.

## <b>Yearly-Month Wise Sales Trend<b>

```python
# my=data.groupby('Year')['Month'].sum()
month_yearly = data.groupby('Month_Year')['Total Revenue'].sum()
```

```python
plt.figure(figsize=(25,10))

sns.lineplot(x=month_yearly.index, y=month_yearly.values, color='red',
             marker='o')
plt.title("Monthly-Yearly Wise Sales Trend")
plt.xlabel("Month-Year")
plt.ylabel("Sales")
plt.xticks(month_yearly.index, rotation=90)
# plt.grid(True)
plt.show()
```

* The company collects the highest revenue July,2013
* And the lowest in 06-2011, 09-2014, 08-2015



## Region wise Total Revenue and Profit

```python
region= data.groupby('Region')['Total Revenue'].sum()
region_profit=data.groupby('Region')['Total Profit'].sum()
```

```python
plt.figure(figsize=(15,6))
# For Total Revenue
plt.subplot(1,2,1)
plt.pie(region, labels=region.index, autopct='%0.1f%%')
plt.title('Region wise Total Revenue')

# For Total Profit
plt.subplot(1,2,2)
sns.barplot(x=region_profit.index, y=region_profit.values,color='red')
plt.title('Region wise Total Profit')
plt.xlabel('regions')
plt.ylabel('Total Profit')
plt.xticks(rotation=90)
plt.show()
```

* Sub-Saharan Region has the highest revenue and the highest Profit followed by Europe
* North America has the lowest Revenue as well as Profit

## Total Revenue on Different Item types

```python
item= data.groupby('Item Type')['Total Revenue'].sum()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x=item.index, y=item.values,color='red')
plt.title('Item types based on Total Revenue')
plt.xlabel('Item Types')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.show()
```

* Upon analyzing the above figure, it is evident that Amazon generates the highest revenue from the Cosmetics items followed by household and office supplies items

## Total Profit on Different Item types

```python
item_profit= data.groupby('Item Type')['Total Profit'].sum()
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(x=item_profit.index, y=item_profit.values, color='blue',
             marker='o')
plt.title("Item types based on Total Profit")
plt.xlabel("Item Types")
plt.ylabel("Total Revenue")
plt.xticks( rotation=45)
# plt.grid(True)
plt.show()
```

* Company made a large amount of prfit on selling Cosmetics Items

## Total Revenue and Profit by sales channel by Sales Channel

```python
revenue_channel= data.groupby('Sales Channel')['Total Revenue'].sum()
profit_channel= data.groupby('Sales Channel')['Total Profit'].sum()
```

```python
plt.figure(figsize=(15,6))
plt.subplot(1,2,1)

plt.pie(revenue_channel, labels=revenue_channel.index, autopct='%0.1f%%')
plt.title('Total Revenue by Sales Channel')

plt.subplot(1,2,2)

plt.pie(profit_channel, labels=profit_channel.index, autopct='%0.1f%%')
plt.title('Total Profit by Sales Channel')
plt.show()
```

* Offline channel has the large number of Revenue as well as Profit
